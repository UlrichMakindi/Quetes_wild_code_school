## Recommending system

```python
"""Qu'est-ce qu'un système de recommandation ? Un système de recommandation est un système qui prédit la note 
ou la préférence qu'un utilisateur donnerait à un élément. Les systèmes de recommandation sont utilisés dans 
divers domaines, notamment les films, la musique, les actualités, les livres, les articles de recherche, 
les requêtes de recherche, les tags sociaux et les produits en général. Il existe deux principaux types de systèmes de recommandation : 
le filtrage collaboratif et les systèmes basés sur le contenu.
 Le filtrage collaboratif est basé sur l'hypothèse selon laquelle les personnes qui ont été d'accord par le passé seront également d'accord à l'avenir. 
 Les systèmes basés sur le contenu sont basés sur l'idée que si vous aimez un élément, vous aimerez également un élément "similaire"."""
```

### Content-Based Filtering Method

```python
""" Exigence en matière de données :
Le filtrage basé sur le contenu repose principalement sur les attributs des éléments et les 
préférences des utilisateurs, ce qui le rend adapté aux scénarios où les caractéristiques explicites des éléments sont disponibles.
Le filtrage collaboratif, en revanche, repose sur les interactions ou les évaluations des utilisateurs avec les éléments,
 ce qui le rend adapté aux scénarios où les caractéristiques explicites des éléments peuvent ne pas être disponibles ou fiables.

Indépendance de l'utilisateur :
Le filtrage basé sur le contenu ne nécessite pas d'informations sur les préférences ou le comportement d'autres utilisateurs. 
Il se concentre uniquement sur les interactions et les préférences individuelles des utilisateurs.
Le filtrage collaboratif repose sur les similitudes entre les utilisateurs ou les éléments en fonction de leurs interactions avec le système, 
ce qui le rend plus dépendant du comportement de plusieurs utilisateurs.

Problème de démarrage à froid :
Le filtrage basé sur le contenu peut mieux gérer le "problème de démarrage à froid" que le filtrage collaboratif pour les nouveaux utilisateurs ou 
les nouveaux éléments, car il peut faire des recommandations uniquement sur la base des attributs des éléments.
Le filtrage collaboratif peut avoir du mal à faire des recommandations précises pour les nouveaux utilisateurs ou les nouveaux éléments 
avec des données d'interaction limitées.

Sérénité vs. Précision :
Le filtrage basé sur le contenu a tendance à fournir des recommandations plus alignées sur les préférences connues de l'utilisateur, 
car il repose sur des caractéristiques explicites des éléments.
Le filtrage collaboratif peut potentiellement offrir des recommandations plus inattendues en identifiant des schémas dans les interactions 
utilisateur-élément qui peuvent ne pas être évidents à partir des caractéristiques des éléments seuls."""
```

## Pipeline Preparation

```python
numeric_features = ['Weight', 'Length_inches', 'Width_inches', 'Height_inches', 'Battery_Life']
category_features = ['Category']
boolean_features = ['Bluetooth']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from category_encoders import BinaryEncoder
```

```python
# Define the preprocessing steps for each feature type
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),  # Impute missing values with median
    ('scaler', StandardScaler())  # Standard scaling
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),  # Impute missing values with 'missing'
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  # One-hot encoding
])

binary_transformer = Pipeline(steps=[
    ('binary', BinaryEncoder())  # Binary encoding
])

# Combine preprocessing steps using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, category_features),
        ('bin', binary_transformer, boolean_features),
    ])
```

#### KNN Recommendations

```python
from sklearn.neighbors import NearestNeighbors

# Define the KNN model
knn_model = NearestNeighbors(n_neighbors=4, algorithm="auto")

# Combine preprocessing and KNN model using Pipeline
pipeline_with_knn = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('knn', knn_model)
])
```

```python
pipeline_with_knn.fit(products_df_cleaned)
```

```python
def find_nearest_neighbors(product_id : str, pipeline_with_knn : Pipeline, products_df : pd.DataFrame) -> pd.DataFrame :
    """Args : product_id : str : The Product_ID of the query product
              pipeline_with_knn : Pipeline : The pipeline containing the preprocessor and KNN model
              products_df : pd.DataFrame : The DataFrame containing the product information
       Returns : pd.DataFrame : A DataFrame containing the nearest neighbors of the query product"""

    # Filter the DataFrame to get the features of the specified product
    query_product_features = products_df[products_df['Product_ID'] == product_id].drop(columns=['Product_ID', 'Product_Name'])

    # Use the pipeline to preprocess the query product features
    query_product_features_processed = pipeline_with_knn.named_steps['preprocessor'].transform(query_product_features)

    # Use the KNN model to find the nearest neighbors for the query product
    nearest_neighbors_indices = pipeline_with_knn.named_steps['knn'].kneighbors(query_product_features_processed)[1][0]

    # Get the nearest neighbors' Product_IDs
    nearest_neighbors_product_ids = products_df.iloc[nearest_neighbors_indices]['Product_ID']

    # Create a DataFrame containing the nearest neighbors' information
    nearest_neighbors_df = products_df[products_df['Product_ID'].isin(nearest_neighbors_product_ids)]

    return nearest_neighbors_df[nearest_neighbors_df['Product_ID'] != product_id]

# Example usage:
query_product_id = 'P1'  # Specify the Product_ID of the query product
nearest_neighbors_df = find_nearest_neighbors(query_product_id, pipeline_with_knn, products_df_cleaned)

nearest_neighbors_df
```

#### Cosine Similarity 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Compute cosine similarity matrix for the selected features
similarity_matrix = cosine_similarity(pipeline_with_knn.named_steps['preprocessor'].transform(products_df_cleaned))

# Convert the similarity matrix to a DataFrame for easier manipulation
similarity_df = pd.DataFrame(similarity_matrix, index=products_df_cleaned['Product_ID'], columns=products_df_cleaned['Product_ID'])
```

```python
similarity_df
```

```python
# Function to get top N similar products for a given product
def get_similar_products(product_id : str, top_n : int =3) -> pd.DataFrame:
    """Args : product_id : str : The Product_ID of the query product
              top_n : int : The number of similar products to return
       Returns : pd.DataFrame : A DataFrame containing the top N similar products to the query product"""
       
    # Get the similarity scores for the given product
    similar_scores = similarity_df.loc[product_id].sort_values(ascending=False)
    # Exclude the given product itself
    similar_scores = similar_scores.drop(product_id)
    # Get the top N similar products
    top_similar_products = similar_scores.head(top_n)    
    
    top_similar_products_df = products_df_cleaned[products_df_cleaned['Product_ID'].isin(top_similar_products.index)]

    return top_similar_products_df
```

```python
# Example usage: Get top 3 similar products for product with ID 'P1'
similar_products = get_similar_products('P1')
similar_products
```

## Collaborative Filtering

```python

"""Bien sûr, voici un texte similaire pour le filtrage collaboratif :

Exigences en matière de données :

Le filtrage collaboratif s'appuie principalement sur les interactions ou les évaluations des utilisateurs avec les éléments,
ce qui le rend adapté aux scénarios où les informations sur les préférences individuelles des utilisateurs sont disponibles.
Contrairement au filtrage basé sur le contenu, il ne dépend pas des caractéristiques explicites des éléments,
ce qui le rend plus flexible pour les cas où ces caractéristiques ne sont pas disponibles ou fiables.

Indépendance de l'utilisateur :

Contrairement au filtrage basé sur le contenu, le filtrage collaboratif nécessite des informations sur les interactions des utilisateurs avec 
les éléments et la similarité entre les utilisateurs en fonction de ces interactions.
Il est donc plus dépendant du comportement de plusieurs utilisateurs et peut fournir des recommandations basées sur les comportements similaires
d'autres utilisateurs.

Problème de démarrage à froid :

Le filtrage collaboratif peut également rencontrer des difficultés avec le "problème de démarrage à froid",
en particulier pour les nouveaux utilisateurs ou les nouveaux éléments,
car il repose sur les interactions utilisateur-élément pour générer des recommandations.
Cependant, certaines approches, telles que la recommandation basée sur des voisins proches,
peuvent atténuer ce problème en identifiant des utilisateurs similaires pour les nouveaux utilisateurs.

Sérénité vs. Précision :

Le filtrage collaboratif a le potentiel de fournir des recommandations plus diversifiées et inattendues en identifiant des schémas dans 
les interactions utilisateur-élément. Cela peut conduire à des recommandations qui ne sont pas strictement basées 
sur les préférences connues de l'utilisateur, mais qui peuvent néanmoins être appréciées.
Cependant, la précision des recommandations dépend fortement de la qualité et de la quantité des données d'interaction disponibles."""
```

```python
import numpy as np
import pandas as pd

# Define the number of users and products
num_users = 20
num_products = 10

# Generate random user IDs
user_ids = np.arange(1, num_users + 1)

# Generate ratings for each user
ratings_data = []
for user_id in user_ids:
    # Determine the number of products to be rated by the user (between 3 and 6)
    num_rated_products = np.random.randint(3, 7)
    
    # Randomly select products to be rated by the user
    products_to_rate = np.random.choice(products_df_cleaned['Product_ID'], size=num_rated_products, replace=False)
    
    # Generate ratings for each product (assign -1 if not rated)
    for product_id in products_df_cleaned['Product_ID']:
        if product_id in products_to_rate:
            # Generate a random rating between 1 and 5 for rated products
            rating = np.random.randint(1, 6)
        else:
            # Assign -1 for products not rated by the user
            rating = -1
        ratings_data.append({'User_ID': user_id, 'Product_ID': product_id, 'Rating': rating})

# Create a DataFrame from the generated ratings data
ratings_df = pd.DataFrame(ratings_data)

# Display the first few rows of the ratings DataFrame
ratings_df.head()
```

```python
#Display of a user - product mtrix in DataFrame format
ratings_matrix = ratings_df.pivot(index='User_ID', columns='Product_ID', values='Rating')
ratings_matrix
```

```python
from surprise import Dataset
from surprise import Reader

# Load the dataset
reader = Reader(rating_scale=(1, 5))
data_surprise = Dataset.load_from_df(ratings_df[['User_ID', 'Product_ID', 'Rating']], reader)
```

```python
from surprise import SVD
from surprise.model_selection import cross_validate

# Define the algorithm
model = SVD(n_factors=10)
```

```python
from surprise.model_selection import train_test_split
from surprise import accuracy

# Split the data into train and test sets
trainset, testset = train_test_split(data_surprise, test_size=0.2)

# Train the modelrithm on the train set
model.fit(trainset)

# Make predictions on the test set
predictions = model.test(testset)
```

```python
predictions
```

```python
# Evaluate the predictions
rmse = accuracy.rmse(predictions)
mae = accuracy.mae(predictions)
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
```

```python
train_predictions = model.test(trainset.build_testset())
```

```python
all_predictions = predictions + train_predictions
```

```python
def recommend_top_unrated_products(all_predictions: list[float], customer_id : int) -> list:
    """Args : all_predictions : list of predictions, customer_id : int
    Returns : list of top 3 unrated product IDs for the specified customer

    Documetation : This function takes a list of predictions and a customer ID as input and returns the 
    top 3 unrated product IDs for the specified customer.
    """
    # Filter predictions for the specified customer where actual rating is -1
    unrated_predictions = [pred for pred in all_predictions if pred.uid == customer_id and pred.r_ui == -1]

    # Sort unrated predictions by estimated rating (est) in descending order
    sorted_predictions = sorted(unrated_predictions, key=lambda x: x.est, reverse=True)

    # Extract the top 3 products that are not already rated by the customer
    top_3_unrated_products = [pred.iid for pred in sorted_predictions][:3]

    return top_3_unrated_products

# Example usage:
customer_id = 1  # Specify the customer ID
top_3_products = recommend_top_unrated_products(all_predictions, customer_id)
print(f"Top 3 unrated products for Customer {customer_id}: {top_3_products}")
```

#### Hybrid filtering (Collaborative + Content)

```python
num_users = 20
num_products = 10

# Generate random user IDs and movie IDs
user_ids = np.arange(1, num_users + 1)
products_ids = np.arange(1, num_products + 1)

# Generate random values indicating whether a user has seen a movie (1) or not (0)
user_item_data = np.random.randint(2, size=(num_users, num_products))

# Create DataFrame from the generated data
user_item_df = pd.DataFrame(user_item_data, index=user_ids, columns=products_ids)

#Create an index names user_id
user_item_df.index.name = 'User_ID'

user_item_df
```

```python
# Step 1: User-Based Collaborative Filtering
# Compute cosine similarity between users based on their purchase behavior
user_similarity = cosine_similarity(user_item_df)

# Identify the 5 closest users to the target user
target_user_index = 0  # Example: Assuming we are recommending for the first user
closest_users_indices = np.argsort(user_similarity[target_user_index])[::-1][1:6]  # Exclude the target user and select top 5
closest_users_similarity = user_similarity[target_user_index, closest_users_indices]

# Aggregate purchase behavior of closest users to generate initial recommendations
initial_recommendations = user_item_df.iloc[closest_users_indices].mean(axis=0)
initial_recommendations = initial_recommendations.where(initial_recommendations > 0, 0)  
```

```python
initial_recommendations
```

```python
similarity_matrix
```

```python
""" Refine recommendations by incorporating similar items -> Le produit matriciel entre les recommandations initiales et la matrice de similarité
permet de pondérer les recommandations initiales en fonction de la similarité entre les produits. Cela permet d'identifier des produits similaires
qui n'ont pas été achetés par l'utilisateur cible, mais qui sont appréciés par des utilisateurs similaires."""
refined_recommendations = initial_recommendations.dot(similarity_matrix)
```

```python
# Step 3: Recommendation Generation
# Combine initial recommendations with refined recommendations
final_recommendations = initial_recommendations + refined_recommendations

# Print final recommendations
print("Final Recommendations:")
print(final_recommendations)
```

```python
target_user_index = 1  # Example: Assuming we are recommending for the first user
# Step 1: Filter out products already bought by the user
products_already_bought = set(user_item_df.columns[user_item_df.loc[target_user_index] > 0])

# Step 2: Sort remaining products based on final recommendation scores
remaining_products = final_recommendations.drop(products_already_bought)
recommended_products_sorted = remaining_products.sort_values(ascending=False)

# Step 3: Select top-n products to recommend
top_n = 3  # Number of top products to recommend
top_recommendations = recommended_products_sorted.head(top_n)

# Print top recommendations
print("Top Recommendations:")
print(top_recommendations)
```
