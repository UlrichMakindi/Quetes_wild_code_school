```python
import json
import requests
import pandas as pd
```

```python
link = 'https://raw.githubusercontent.com/murpi/wilddata/master/quests/monthlySalesbyCategoryMultiple.json'
r = requests.get(link)

r.text
data = json.loads(r.text)
data
```

```python
pd.json_normalize(

    data['contents'],
    record_path='monthlySales',
    meta = ['category', 'region']
    )
```
