```python
import pandas as pd
import requests
```

```python
link = 'https://data.nantesmetropole.fr/api/explore/v2.1/catalog/datasets/244400404_stations-velos-libre-service-nantes-metropole-disponibilites/records?select=available_bikes%2C%20address%2C%20bike_stands&limit=20'
```

```python
r = requests.get(link)
print(r.json())
```

```python
data = pd.json_normalize(r.json(),
                         record_path='results',
                         )
```

```python
data['taux_de_remplissage'] = data['available_bikes'] / data['bike_stands']
```

```python
data.drop(columns=['available_bikes', 'bike_stands'], inplace=True)
```
